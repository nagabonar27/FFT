sine waves with y(t) = Asin(wt+phi)

```python
samplingrate = 9510 #hz
interval = 1/samplingrate
time = np.arange(0,1,interval)

f = 3

y = np.sin(2*np.pi*f*time)

pl.ylabel('amplitude')
pl.xlabel('time')
pl.plot(time,y)
```

```python
yfft = fft(y) #calculate the fft of y
N = len(yfft)
n = np.arange(N)
T = N/samplingrate
freq = n/T

#plt.figure(figsize = (12, 6))

pl.subplot(121)
pl.xlabel('freq (Hz)')
pl.ylabel('FFT Amplitude |X(freq)|')
pl.stem(freq, np.abs(yfft), 'b', \
         markerfmt=" ", basefmt="-b")
pl.xlim(0,10)

pl.subplot(122)
pl.plot(time,y)
pl.xlabel('time')
pl.ylabel('amplitude')
pl.tight_layout

pl.show()
```

```python
fig, (axis1, axis2) = pl.subplots(2,1)

axis1.set_subplotspec
axis1.set_xlabel('freq(HZ)')
axis1.set_ylabel('FFT Amplitude |X(freq)|')
axis1.stem(freq, np.abs(yfft), 'b', \
         markerfmt=" ", basefmt="-b")

axis2.plot(time,y)
```
