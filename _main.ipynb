```python
x = np.linspace(0, 10,  ) # create a location using linspace
y = np.sin(x)

```

```python
samplingrate = 800 #hz
interval = 1/samplingrate
time = np.arange(0,1,interval)

f = 5

y = np.sin(2*np.pi*f*time)

#noise addition
noise_sample = 800
y_noise = y + 0.008* np.asarray(random.sample(range(0,1000),noise_sample))


fig,ax = pl.subplots(1,2,figsize = (10,2))
fig.tight_layout


ax[0].plot(time,y, color='red')
ax[0].set_xlabel('before noise')
ax[1].plot(time,y_noise, color='blue')
ax[1].set_xlabel('after noise')
pl.show
```

```python
yfft = fft(y) #calculate the fft of y
yfft_noise = fft(y_noise)
N = len(yfft)
n = np.arange(N)
T = N/samplingrate
freq = n/T

#plt.figure(figsize = (12, 6))


pl.xlabel('freq ')
pl.ylabel('amp')
pl.stem(freq, np.abs(yfft), 'b', \
         markerfmt=" 1", basefmt="-b")
pl.xlim(0,10)

```
